```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

## BAR CHART

```python
matakuliah = ['Dasar Pemrograman','Basis Data','Machine Learning','Pemrograman Objek','Pemrograman Web']
jumlahmahasiswa = [100, 70, 150, 90,30]

plt.figure(figsize=(12,7))
plt.barh(matakuliah, jumlahmahasiswa, color='blue')
plt.title('Diagram Batang Mata Kuliah Informatika', size=16)
plt.xlabel('Jumlah Mahasiswa')
plt.show()
```

## LINE CHART (DIAGRAM GARIS)

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
nilai = [80,95,99,50,75]
ax.plot(nilai)
plt.show()
```

## PIE CHART

```python
fig = plt.figure()
produk = ['Cheese', 'Roti', 'Milk', 'Beras', 'Jagung'] 
sold = [20,40,25,10,12]
ax = fig.add_axes([0,0,1,1])
ax.pie(sold, labels=produk, autopct='%1.2f%%')
plt.show()
```

## MENGATASI DATA KOSONG (MISSING VALUE)

```python
# MODUS untuk data yang kategorik
# MEAN untuk data numerik

from statistics import mode
jenis_kelamin_mode = mode(df['jenis_kelamin'])
pekerjaan_mode = mode(df['pekerjaan'])
is_keturunan_mode = mode(df['is_keturunan'])
sampo_mode = mode(df['sampo'])
berat_mean = df['berat'].mean()
gaji_mean = df['gaji'].mean()

kolom_missing = ['jenis_kelamin', 'pekerjaan'
                'is_keturunan', 'berat', 'sampo', 'gaji']
inputan = [jenis_kelamin_mode, pekerjaan_mode, 
          is_keturunan_mode, berat_mean, sampo_mode, gaji_mean]

df['jenis_kelamin'].fillna(jenis_kelamin_mode, inplace=True)
```
